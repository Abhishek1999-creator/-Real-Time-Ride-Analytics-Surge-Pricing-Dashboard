```python
import random
import time
import psycopg2
import pandas as pd

# 🔌 Connect to PostgreSQL        
conn = psycopg2.connect(
    host="localhost",
    database="rides_db",
    user="postgres",
    password="1319",
    port="5432"
)

cursor = conn.cursor()

cities = ["Kolkata", "Delhi", "Mumbai"]

print("✅ Streaming to PostgreSQL started...")

try:
    while True:
        
        status = random.choices(
            ["completed", "cancelled"],
            weights=[80, 20]
        )[0]

       
        if status == "completed":
            fare = random.randint(150, 1000)
        else:
            fare = random.randint(0, 300)

        data = (
            random.randint(1000, 9999),   # ride_id
            random.choice(cities),        # city
            fare,                         # fare
            status,                       # status
            pd.Timestamp.now()            # timestamp
        )

        cursor.execute("""
            INSERT INTO rides (ride_id, city, fare, status, timestamp)
            VALUES (%s, %s, %s, %s, %s)
        """, data)

        conn.commit()

        print("Inserted:", data)

        time.sleep(2)

except KeyboardInterrupt:
    print("🛑 Streaming stopped")

finally:
    cursor.close()
    conn.close()
```
